```python
t=table.__str__()
t[:1000]
```

```python
t=t.replace(',','.')
t[:1000]
```

```python
#pasa los datos a tabla de python
df=pandas.read_html(t, header=0)[0]
df.head()
```

```python
d=df.to_dict(orient="split")
```

```python
d.keys()
```

```python
data=d['data']
data
```

```python
gc = geocoder.google("Uganda")
print(gc)
```

```python
gc.latlng
```

```python
l=[]
for nombre, tasa, a√±o in data:
    lugar= nombre +", "+str(tasa)+", "+str(a√±o) #se escribe str(7658) para  hacerlo string porque "nombre" es string pero "tasa" es un nro entonces lo transformo a string
    print(lugar)
    latlng= geocoder.google(lugar).latlng
    print(latlng)
    pais= [nombre, tasa, a√±o, latlng]
    l.append(pais)
    
```

```python
l
```

```python
len(l)
```

```python
l=[[nombre, tasa, a√±o, latlng] for nombre, tasa, a√±o, latlng in l if latlng != None] #si sale none (que no tiene latitud) que se borre
```

```python
len(l)
```

```python
m=folium.Map(
    location = [39.39987199999999, -8.224454],
    zoom_start=1
)
for nombre, tasa, a√±o, latlng in l:
    folium.Marker(latlng).add_to(m)
m
```

```python
m=folium.Map(
    location = [39.39987199999999, -8.224454],
    zoom_start=1
)
for nombre, tasa, a√±o, latlng in l:
    folium.Circle(
        latlng,
        weight=0,
        color='green',
        fill=True,
        radius = 15000*tasa
    ).add_to(m)
m
```

```python
import folium
import pandas as pd

state_geo = r'us-states.json'
state_unemployment = r'US_Unemployment_Oct2012.csv'

state_data = pd.read_csv(state_unemployment)

#Let Folium determine the scale
map = folium.Map(location=[48, -102], zoom_start=3)
map.geo_json(geo_path=state_geo, data=state_data,
             columns=['State', 'Unemployment'],
             key_on='feature.id',
             fill_color='YlGn', fill_opacity=0.7, line_opacity=0.2,
             legend_name='Unemployment Rate (%)')
map
```
